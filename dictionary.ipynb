```python
from pdfreader import PDFDocument
from pdfreader import SimplePDFViewer

fd = open('Oxford_Dictionary_of_Idioms.pdf', "rb")
doc = PDFDocument(fd)
viewer = SimplePDFViewer(fd)
# page = next(doc.pages())


# print(page.Resources)
# sorted(page.Resources.Font.keys())

# font = page.Resources.Font['T1_0']
# font.Subtype, font.BaseFont, font.Encoding
```

```python
for idx, page in enumerate(doc.pages()):
    if (idx == 14):
        for keys in page.Resources.Font.keys():
            font = page.Resources.Font[keys]
            print(font.BaseFont)
        
        viewer.navigate(idx)
        viewer.render()
        strings = viewer.canvas.text_content
        print(strings)

```
